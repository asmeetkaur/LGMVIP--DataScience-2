```python
import cv2
import numpy as np

def image_to_pencil_sketch(image_path, output_path):
    # Read the image in RGB format
    original_image = cv2.imread(image_path)
    rgb_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)

    # Invert the grayscale image
    inverted_image = cv2.bitwise_not(grayscale_image)

    # Blur the inverted image to create an inverted blurry image
    blurred_image = cv2.GaussianBlur(inverted_image, (111, 111), 0)

    # Invert the blurred image
    inverted_blurred_image = cv2.bitwise_not(blurred_image)

    # Create the pencil sketch by dividing the grayscale image by the inverted blurry image
    pencil_sketch = cv2.divide(grayscale_image, inverted_blurred_image, scale=256.0)

    # Save the pencil sketch
    cv2.imwrite(output_path, pencil_sketch)

if __name__ == "__main__":
    # Provide the path to your input image
    input_image_path = "C:\\Users\\lenovo\\OneDrive\\Desktop\\image.jpg"

    # Provide the desired path for the output pencil sketch image
    output_sketch_path = "C:\\Users\\lenovo\\OneDrive\\Desktop\\output_sketch.jpg"

    # Convert the image to a pencil sketch
    image_to_pencil_sketch(input_image_path, output_sketch_path)
```
